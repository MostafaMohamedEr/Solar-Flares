# Space weather live page

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime
from bs4 import BeautifulSoup
#Getting page Text
spaceWeatherPage=requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares')
spaceWeatherPageText=spaceWeatherPage.text
spaceWeatherPageText[:1000]
```

```python
#Using beautiful soup to organize Html Text
page=BeautifulSoup(spaceWeatherPageText)
page.prettify()[:1000]
```

```python
#Reterving the table from the page
Table=page.find('table','table table-striped table-responsive-md')
#Defining lists that will be used as columns in data frame
rank=[]
x_class=[]
date=[]
region=[]
start_time=[]
max_time=[]
end_time=[]
movie=[]
"""""
After Reterving the table, rows will be stored in variable,
then iterate in every row to get each variable to append it to it's list(each iteration we add 8 to go to next row)
"""""
Rows=Table.find_all('td')
for rowIndex in range(len(Table.find_all('tr'))-1):
    rank.append(Rows[0+8*rowIndex].text)
    x_class.append(Rows[1+8*rowIndex].text)
    date.append(Rows[2+8*rowIndex].text)
    region.append(Rows[3+8*rowIndex].text)
    start_time.append(Rows[4+8*rowIndex].text)
    max_time.append(Rows[5+8*rowIndex].text)
    end_time.append(Rows[6+8*rowIndex].text)
    movie.append(Rows[7+8*rowIndex].text)
spaceWeatherDataFrame=pd.DataFrame({'Rank':rank,'X_class':x_class,'Date':date,'Region':region,'Start_time':start_time,'Max_time':max_time,'End_time':end_time,'Movie':movie})
spaceWeatherDataFrame
```

```python
modifiedDataFrame=spaceWeatherDataFrame
modifiedDataFrame.drop(['Movie'],axis=1,inplace=True)
modifiedDataFrame.head()
```

```python
modifiedDataFrame['Start_time']=modifiedDataFrame['Date']+" "+modifiedDataFrame['Start_time']
modifiedDataFrame['End_time']=modifiedDataFrame['Date']+" "+modifiedDataFrame['End_time']
modifiedDataFrame.drop(['Date'],axis=1,inplace=True)
modifiedDataFrame.head()
```

```python
modifiedDataFrame['Start_time']=pd.to_datetime(modifiedDataFrame['Start_time'])
modifiedDataFrame['End_time']=pd.to_datetime(modifiedDataFrame['End_time'])
modifiedDataFrame[:1000]
```
